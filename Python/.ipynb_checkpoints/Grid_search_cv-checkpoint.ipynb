```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('species',axis=1)
y = df['species']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=17)
```

```python
model_knn.fit(x_train,y_train)
```

```python
model_knn.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma='auto')
```

```python
model_svm.fit(x_train,y_train)
```

```python
model_svm.score(x_train,y_train)
```

#NoW leatest Use the Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifire = GridSearchCV(model_svm, { 
    'C': [1,10,20,30],
    'kernel': ['rbf','linear'],
}, cv=5, return_train_score=False)
```

```python
classifire.fit(x,y)
```

```python
classifire.cv_results_
```

```python
results = pd.DataFrame(classifire.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['minkowski', 'euclidean']
}
```

```python
classifire_knn = GridSearchCV(model_knn, param_grid, cv=5, return_train_score=False)
```

```python
classifire_knn.fit(x, y) 
```

```python
results_knn = pd.DataFrame(classifire_knn.cv_results_)
```

```python
results_knn
```

```python
print(results_knn[['param_n_neighbors', 'param_weights', 'mean_test_score']])
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifire_random = RandomizedSearchCV(model_svm, { 
    'C': [1,10,20,30],
    'kernel': ['rbf','linear'],
},n_iter=4, cv=5, return_train_score=False)
```

```python
classifire_random.fit(x,y)
```

```python
results = pd.DataFrame(classifire_random.cv_results_)
```

```python
results[['param_C','param_kernel','mean_test_score']]
```
